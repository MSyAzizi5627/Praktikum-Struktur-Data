Representasi Graf dengan Adjacency List

```python
# Membuat graf tak berarah menggunakan adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}


# Menampilkan graf
def print_graph(graf):
    for node in graf:
        print(f"{node} -> {graf[node]}")


print_graph(graph)
```

Traversal Graf menggunakan BFS

```python
from collections import deque


def bfs(graph, start):
    visited = set()
    queue = deque([start])


    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)
            queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])


# Contoh pemanggilan
bfs(graph, 'A')

```

Traversal Graf menggunakan DFS

```python
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()


    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)


# Contoh pemanggilan
dfs(graph, 'A')
```

```python
def count_neighbors(graph):
    result = {}
    for node in graph:
        # Menghitung jumlah elemen dalam list tetangga untuk node tersebut
        result[node] = len(graph[node])
    return result

jumlah_tetangga = {
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A'],
    'D': ['B']
}

print(count_neighbors(jumlah_tetangga))
```

```python
from collections import deque

# 2&3. BFS (Breadth-First Search) menggunakan Queue
def bfs_visited(graph, start):
    visited = []
    queue = deque([start])

    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.append(node)
            # Tambahkan tetangga yang belum dikunjungi ke antrean
            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)
    return visited

# DFS (Depth-First Search) menggunakan Rekursi
def dfs_visited(graph, start, visited=None):
    if visited is None:
        visited = []

    visited.append(start)
    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs_visited(graph, neighbor, visited)
    return visited

print("\nUrutan BFS dari A:", bfs_visited(my_graph, 'A'))
print("Urutan DFS dari A:", dfs_visited(my_graph, 'A'))
```

```python
# 4. fungsi find_path
def find_path(graph, start, end, path=None):
    if path is None:
        path = []

    path = path + [start]

    if start == end:
        return path

    for neighbor in graph[start]:
        if neighbor not in path:
            new_path = find_path(graph, neighbor, end, path)
            if new_path:
                return new_path
    return None

print("\nJalur dari A ke D:", find_path(my_graph, 'A', 'D'))
```

```python
# 5. Menambahkan is_connected
def is_connected(graph):
    if not graph:
        return True

    # Ambil salah satu simpul sebagai titik mulai
    start_node = list(graph.keys())[0]
    visited_nodes = dfs_visited(graph, start_node)

    # Bandingkan jumlah node yang berhasil dikunjungi dengan total node
    return len(visited_nodes) == len(graph)

print("\nApakah graf saling terhubung?", is_connected(my_graph))
```

```python
# 1. Membuat Graf
my_graph = {
    'A': ['B', 'C', 'E'],
    'B': ['A', 'C', 'D'],
    'C': ['A', 'B', 'D', 'E'],
    'D': ['B', 'C', 'E'],
    'E': ['A', 'C', 'D']
}

def print_graph(graph):
    for node, neighbors in graph.items():
        print(f"{node}: {', '.join(neighbors)}")

print("Struktur Graf:")
print_graph(my_graph)

# 2.Implementasikan BFS dan DFS
# 3. Modifikasi fungsi BFS agar mengembalikan list urutan kunjungan
from collections import deque

# BFS (Breadth-First Search) menggunakan Queue
def bfs_visited(graph, start):
    visited = []
    queue = deque([start])

    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.append(node)
            # Tambahkan tetangga yang belum dikunjungi ke antrean
            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)
    return visited

# DFS (Depth-First Search) menggunakan Rekursi
def dfs_visited(graph, start, visited=None):
    if visited is None:
        visited = []

    visited.append(start)
    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs_visited(graph, neighbor, visited)
    return visited

print("\nUrutan BFS dari A:", bfs_visited(my_graph, 'A'))
print("Urutan DFS dari A:", dfs_visited(my_graph, 'A'))

# 4. Buat fungsi find_path
def find_path(graph, start, end, path=None):
    if path is None:
        path = []

    path = path + [start]

    if start == end:
        return path

    for neighbor in graph[start]:
        if neighbor not in path:
            new_path = find_path(graph, neighbor, end, path)
            if new_path:
                return new_path
    return None

print("\nJalur dari A ke D:", find_path(my_graph, 'A', 'D'))

# 5. Menambahkan fungsi is_connected
def is_connected(graph):
    if not graph:
        return True

    # Ambil salah satu simpul sebagai titik mulai
    start_node = list(graph.keys())[0]
    visited_nodes = dfs_visited(graph, start_node)

    # Bandingkan jumlah node yang berhasil dikunjungi dengan total node
    return len(visited_nodes) == len(graph)

print("\nApakah graf saling terhubung?", is_connected(my_graph))
```
