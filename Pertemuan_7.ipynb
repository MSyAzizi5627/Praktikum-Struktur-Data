Binary Tree Manual

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

# Membuat tree secara manual
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

# Fungsi inorder traversal
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)

print("Hasil inorder traversal dari binary tree:")
inorder(root)
```

Binary Search Tree

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, root, data):
        if root is None:
            return Node(data)
        if data < root.data:
            root.left = self.insert(root.left, data)
        else:
            root.right = self.insert(root.right, data)
        return root

bst = BinarySearchTree()
root = None
data_list = [50, 30, 70, 20, 40, 60, 80]
for value in data_list:
    root = bst.insert(root, value)
```

Traversal pada BST

```python
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)

def preorder(node):
    if node:
        print(node.data, end=' ')
        preorder(node.left)
        preorder(node.right)

def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=' ')

print("Inorder Traversal:")
inorder(root)
print("\nPreorder Traversal:")
preorder(root)
print("\nPostorder Traversal:")
postorder(root)
```

Pencarian Nilai dalam BST

```python
def search(node, key):
    if node is None or node.data == key:
        return node
    if key < node.data:
        return search(node.left, key)
    return search(node.right, key)

# Uji pencarian
key = 60
result = search(root, key)
if result:
    print(f"{key} ditemukan dalam tree.")
else:
    print(f"{key} tidak ditemukan.")

key = 25
result = search(root, key)
if result:
    print(f"{key} ditemukan dalam tree.")
else:
    print(f"{key} tidak ditemukan.")
```

Tugas 1 Program Struktur Binary Tree Manual

```python
# Struktur Tree berdasarkan identitas saya

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def inorder_traversal(node):
    if node:
        inorder_traversal(node.left)
        print(node.data, end=' ')
        inorder_traversal(node.right)

# --- BAGIAN DATA (Silakan sesuaikan dengan identitas Anda) ---
nama_lengkap = "Budi Santoso"  # Contoh nama
npm_terakhir = 24              # Contoh 2 digit terakhir NPM
tanggal_lahir = 15             # Contoh tanggal (DD)
bulan_lahir = 10               # Contoh bulan (MM)

# Logika perhitungan manual
jumlah_huruf = len(nama_lengkap.replace(" ", ""))
vokal = "aiueoAIUEO"
jumlah_vokal = sum(1 for char in nama_lengkap if char in vokal)

# --- MEMBANGUN TREE SECARA MANUAL ---

# 1. Node root: jumlah huruf nama lengkap
root = Node(jumlah_huruf)

# 2. Anak kiri root: jumlah huruf vokal
root.left = Node(jumlah_vokal)

# 3. Anak kanan root: dua digit terakhir NPM
root.right = Node(npm_terakhir)

# 4. Tambah satu level lagi:
# Anak kiri dari node kiri: tanggal (DD)
root.left.left = Node(tanggal_lahir)

# Anak kanan dari node kanan: bulan (MM)
root.right.right = Node(bulan_lahir)

# --- OUTPUT ---
print("Struktur Tree ditampilkan menggunakan inorder traversal:")
inorder_traversal(root)
print() # Baris baru
```

Tugas 2 Program Binary Search Tree (BST)

---



```python
# Program Binary Search Tree (BST) Berdasarkan Identitas

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    # Fungsi untuk menyisipkan angka secara otomatis sesuai aturan BST
    def insert(self, root, data):
        if root is None:
            return Node(data)
        if data < root.data:
            root.left = self.insert(root.left, data)
        elif data > root.data: # Menghindari duplikasi angka
            root.right = self.insert(root.right, data)
        return root

    # Fungsi Pencarian
    def search(self, node, key):
        if node is None or node.data == key:
            return node
        if key < node.data:
            return self.search(node.left, key)
        return self.search(node.right, key)

# --- Fungsi Traversal ---
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)

def preorder(node):
    if node:
        print(node.data, end=' ')
        preorder(node.left)
        preorder(node.right)

def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=' ')

# --- Persiapan Data (Silakan sesuaikan data ini) ---
# Contoh: Nama "Budi", NPM "202400123", Lahir "15-10"
tiga_digit_npm = 123                # Aturan 1
dd, mm = 15, 10                     # Aturan 2
ascii_1 = ord('B')                  # Aturan 3 (Huruf pertama 'B')
ascii_2 = ord('u')                  # Aturan 3 (Huruf kedua 'u')
bebas_1, bebas_2 = 99, 7            # Aturan 4

# Membuat list 7 angka unik
data_list = [tiga_digit_npm, dd, mm, ascii_1, ascii_2, bebas_1, bebas_2]

# --- Eksekusi Program ---
bst = BinarySearchTree()
root = None

# Menyisipkan semua angka ke dalam BST
for nilai in data_list:
    root = bst.insert(root, nilai)

print("=== HASIL TRAVERSAL BST ===")
print("Inorder   :", end=' ')
inorder(root)
print("\nPreorder  :", end=' ')
preorder(root)
print("\nPostorder :", end=' ')
postorder(root)

print("\n\n=== FITUR PENCARIAN ===")
# 1. Cari dua digit terakhir NPM (Contoh NPM belakang 23)
key1 = 23
result1 = bst.search(root, key1)
print(f"Pencarian {key1}: {'Ditemukan' if result1 else 'Tidak Ditemukan'}")

# 2. Cari angka yang tidak ada di list
key2 = 999
result2 = bst.search(root, key2)
print(f"Pencarian {key2}: {'Ditemukan' if result2 else 'Tidak Ditemukan'}")
```
