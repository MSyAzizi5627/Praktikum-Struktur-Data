Implementasi Hash Table Sederhana

```python
class SimpleHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        self.table[index] = value

    def search(self, key):
        index = self.hash_function(key)
        return self.table[index]

# Contoh penggunaan
ht = SimpleHashTable(10)
ht.insert("apel", 100)
ht.insert("pisang", 200)

print("Nilai untuk 'apel':", ht.search("apel"))
print("Nilai untuk 'pisang':", ht.search("pisang"))
```

Penanganan Collision dengan Linear Probing

```python
class LinearProbingHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        original_index = index

        while self.table[index] is not None:
            if self.table[index][0] == key:
                break
            index = (index + 1) % self.size
            if index == original_index:
                raise Exception("Hash table penuh")

        self.table[index] = (key, value)

    def search(self, key):
        index = self.hash_function(key)
        original_index = index

        while self.table[index] is not None:
            if self.table[index][0] == key:
                return self.table[index][1]
            index = (index + 1) % self.size
            if index == original_index:
                break

        return None

# Contoh penggunaan
ht = LinearProbingHashTable(5)
ht.insert("apel", 100)
ht.insert("pisang", 200)
ht.insert("melon", 300)

print("Nilai untuk 'apel':", ht.search("apel"))
print("Nilai untuk 'melon':", ht.search("melon"))
```

Penanganan Collision dengan Chaining

```python
class ChainingHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                self.table[index][i] = (key, value)
                return
        self.table[index].append((key, value))

    def search(self, key):
        index = self.hash_function(key)
        for k, v in self.table[index]:
            if k == key:
                return v
        return None

# Contoh penggunaan
ht = ChainingHashTable(5)
ht.insert("apel", 100)
ht.insert("pisang", 200)
ht.insert("melon", 300)

print("Nilai untuk 'apel':", ht.search("apel"))
print("Nilai untuk 'melon':", ht.search("melon"))
```

1. Eksperimen sederhana Linier Probing

```python
class LinearProbingHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        # Menggunakan fungsi bawaan Python untuk simulasi hash
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        original_index = index

        while self.table[index] is not None:
            if self.table[index][0] == key:
                break
            index = (index + 1) % self.size
            if index == original_index:
                raise Exception("Hash table penuh")

        self.table[index] = (key, value)

    def print_table(self):
        print("Isi self.table:")
        for idx, element in enumerate(self.table):
            print(f"Slot {idx}: {element}")

# --- EKSPERIMEN ---
# a. Buat objek LinearProbingHashTable dengan ukuran tabel 5
ht_linear = LinearProbingHashTable(5)

# b. Masukkan pasangan key-value yang diminta
ht_linear.insert("A", 10)
ht_linear.insert("B", 20)
ht_linear.insert("C", 30)
ht_linear.insert("D", 40)
ht_linear.insert("E", 50)

# c. Cetak isi array self.table
ht_linear.print_table()
```

Apa yang terjadi jika dua key memiliki hasil hash yang sama?
Jika dua key menghasilkan indeks hash yang sama (collision), metode Linear Probing akan mencari slot kosong terdekat berikutnya secara berurutan (indeks + 1). Proses pencarian ini akan terus bergeser ke kanan (dan kembali ke indeks 0 jika sudah mencapai ujung tabel) hingga menemukan slot yang bernilai None atau memiliki key yang sama untuk diperbarui

2. Percobaan Dasar Chaining

```python
class ChainingHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)] # Membuat list di dalam list

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                self.table[index][i] = (key, value)
                return
        self.table[index].append((key, value))

    def print_table(self):
        print("Isi self.table (Chaining):")
        for idx, slot in enumerate(self.table):
            print(f"Slot {idx}: {slot}")

# --- PERCOBAAN ---
# a. Gunakan ukuran tabel 3
ht_chaining = ChainingHashTable(3)

# b. Masukkan pasangan key-value yang diminta
ht_chaining.insert("apel", 100)
ht_chaining.insert("melon", 200)
ht_chaining.insert("lemon", 300)
ht_chaining.insert("pisang", 400)

# c. Cetak isi tabel dan tampilkan data di setiap slot
ht_chaining.print_table()
```

Penjelasan (Instruksi d)
Bagaimana collision ditangani dalam metode chaining?
Dalam metode Chaining, setiap slot pada hash table tidak langsung menyimpan satu data tunggal, melainkan sebuah struktur data dinamis (seperti list atau linked list). Ketika terjadi collision (dua atau lebih key memiliki indeks hash yang sama), data baru tersebut tidak mencari tempat di slot lain, melainkan dimasukkan atau digabungkan ke dalam list pada slot yang sama (append). Oleh karena itu, satu slot bisa menyimpan banyak pasangan key-value sekaligus.
